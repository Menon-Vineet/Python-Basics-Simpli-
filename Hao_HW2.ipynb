```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import binom

def plot_binomial_distribution():
    # Ask user for input values
    n = int(input("Enter the number of trials (n): "))
    p = float(input("Enter the probability of success (p) [between 0 and 1]: "))

    # Validate inputs
    if n < 0:
        print("Error: The number of trials n must be a non-negative integer.")
        return
    if not 0 <= p <= 1:
        print("Error: The probability p must be between 0 and 1.")
        return

    # Generate possible number of successes
    k = np.arange(0, n+1)
    # Calculate binomial PMF for each k
    pmf = binom.pmf(k, n, p)

    # Plot the PMF
    plt.figure(figsize=(10, 6))
    plt.bar(k, pmf, color='skyblue', edgecolor='black')
    plt.title(f'Binomial Distribution PMF (n={n}, p={p})')
    plt.xlabel('Number of Successes (k)')
    plt.ylabel('Probability')
    plt.xticks(k)
    plt.grid(axis='y', linestyle='--', alpha=0.7)
    plt.show()

# Call the function
plot_binomial_distribution()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import geom, norm, binom, poisson

def plot_distribution():
    print("Select a distribution to plot:")
    print("1. Geometric")
    print("2. Normal")
    print("3. Binomial")
    print("4. Poisson")

    choice = input("Enter the number corresponding to your choice: ")

    if choice == '1':
        # Geometric Distribution
        p = float(input("Enter the probability of success (p) [between 0 and 1]: "))
        if not 0 < p <= 1:
            print("Error: p must be between 0 and 1 (exclusive of 0).")
            return
        k = np.arange(1, 20)  # Geometric distribution starts from 1
        pmf = geom.pmf(k, p)
        plt.bar(k, pmf, color='skyblue', edgecolor='black')
        plt.title(f'Geometric Distribution PMF (p={p})')
        plt.xlabel('Number of Trials until First Success')
        plt.ylabel('Probability')
        plt.xticks(k)
        plt.grid(axis='y', linestyle='--', alpha=0.7)
        plt.show()

    elif choice == '2':
        # Normal Distribution
        mu = float(input("Enter the mean (μ): "))
        sigma = float(input("Enter the standard deviation (σ): "))
        if sigma <= 0:
            print("Error: Standard deviation σ must be positive.")
            return
        x = np.linspace(mu - 4*sigma, mu + 4*sigma, 1000)
        pdf = norm.pdf(x, mu, sigma)
        plt.plot(x, pdf, color='blue')
        plt.title(f'Normal Distribution PDF (μ={mu}, σ={sigma})')
        plt.xlabel('x')
        plt.ylabel('Probability Density')
        plt.grid(True)
        plt.show()

    elif choice == '3':
        # Binomial Distribution
        n = int(input("Enter the number of trials (n): "))
        p = float(input("Enter the probability of success (p) [between 0 and 1]: "))
        if n < 0:
            print("Error: The number of trials n must be a non-negative integer.")
            return
        if not 0 <= p <= 1:
            print("Error: The probability p must be between 0 and 1.")
            return
        k = np.arange(0, n+1)
        pmf = binom.pmf(k, n, p)
        plt.bar(k, pmf, color='skyblue', edgecolor='black')
        plt.title(f'Binomial Distribution PMF (n={n}, p={p})')
        plt.xlabel('Number of Successes')
        plt.ylabel('Probability')
        plt.xticks(k)
        plt.grid(axis='y', linestyle='--', alpha=0.7)
        plt.show()

    elif choice == '4':
        # Poisson Distribution
        lam = float(input("Enter the average rate of success (λ): "))
        if lam <= 0:
            print("Error: The rate λ must be positive.")
            return
        k_max = int(lam + 4*np.sqrt(lam))
        k = np.arange(0, k_max+1)
        pmf = poisson.pmf(k, lam)
        plt.bar(k, pmf, color='skyblue', edgecolor='black')
        plt.title(f'Poisson Distribution PMF (λ={lam})')
        plt.xlabel('Number of Occurrences')
        plt.ylabel('Probability')
        plt.xticks(k)
        plt.grid(axis='y', linestyle='--', alpha=0.7)
        plt.show()

    else:
        print("Invalid choice. Please select a number between 1 and 4.")

# Run the function
plot_distribution()
```
