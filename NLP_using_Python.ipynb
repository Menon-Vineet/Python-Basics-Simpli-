```python
# Installs and imports
!pip3 install nltk
!pip3 install scikit-learn
!pip3 install gensim
!pip install wordcloud
!pip install sentence-transformers # Install sentence-transformers instead of Document Similarity

import numpy as np
import pandas as pd
import nltk
import string
from nltk.corpus import stopwords
from collections import Counter
from sklearn.feature_extraction.text import TfidfVectorizer
from wordcloud import WordCloud
import matplotlib.pyplot as plt
from gensim.models import Word2Vec
from sentence_transformers import SentenceTransformer, util # Import SentenceTransformer instead of DocumentSimilarity

# Assuming you have two documents you want to compare
doc1 = "This is the first document."
doc2 = "This document is the second document."

# Load a pre-trained model (you can choose from various models)
model = SentenceTransformer('all-mpnet-base-v2')

# Encode the documents into embeddings
embedding1 = model.encode(doc1, convert_to_tensor=True)
embedding2 = model.encode(doc2, convert_to_tensor=True)

# Calculate cosine similarity between embeddings
cosine_scores = util.cos_sim(embedding1, embedding2)

print("Cosine Similarity:", cosine_scores.item())
```
